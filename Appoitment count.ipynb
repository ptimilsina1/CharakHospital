```python
# import requisite libraries 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import datetime

pd.Series.__unicode__ = pd.Series.to_string
pd.set_option('display.max_colwidth', 1000)
pd.set_option('display.max_rows', 1000)
```

```python
data = pd.read_excel("input/AppointmentList2023-01-16.xlsx")
data.head()
```

```python
data['Patient Name'] = data['Patient Name'].str.title()

data['Date/Time'] = data['Date/Time'].astype('str')
yr = []
mt = []
yr_mt = []
yr_mt_dd = []
dd = []
for tt in data['Date/Time'].astype('str'):
    yr.append(tt[0:4])
    mt.append(tt[5:7])
    yr_mt.append(tt[0:4]+'-'+tt[5:7])
    yr_mt_dd.append(tt[0:4]+'-'+tt[5:7]+'-'+tt[8:10])
    dd.append(tt[8:10])
data['Year'] = yr
data['Month'] = mt
data['YearMonth'] = yr_mt
data['Date'] = yr_mt_dd
data['Day'] = dd
data.head()
```

```python
data = data[(data['YearMonth']>='2078-09')&(data['YearMonth']<='2079-09')]
print(max((data['YearMonth'])),min(data['YearMonth']))
```

```python
data_new = data[data['Appointment Type']=='New']
```

```python
df = data_new

result = df.groupby(['Hospital No.']).agg({'Date': lambda x: (sorted(x.tolist()), len(x)), 'Patient Name':'first',
                                           'Phone': 'first','Gender': 'first' }).reset_index()

# split the 'Date_Time (All), Count' column into two separate columns
result[['Date(All)', 'Count']] = pd.DataFrame(result['Date'].tolist(), index=result.index)

# drop the original 'Date_Time' column and rename the 'Date_Time (All)' column
result = result.drop('Date', axis=1).rename(columns={'Date(All)': 'Date_Time'})



result = result.sort_values(['Count'], ascending=False)
result.head(100)
```

```python
# export the resulting data frame to an Excel file
result.to_excel('Patient_visits_count_Poush78toPoush79.xlsx', index=False)
```
